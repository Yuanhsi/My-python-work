```python
df = pd.read_csv("http://bit.ly/uforeports")
```

```python
df.head()
```

```python
df_state = df.groupby("State").count()
```

```python
df_state
```

```python
df_state.sort_values(by="Time", ascending = False)
```

```python
df_state #這樣看發現原始資料沒有重新排序
```

```python
df_state.sort_values(by="Time", ascending = False, inplace = True) #加上inplace 就會改變原始資料排序
```

```python
df_state.head(5)
```

```python
df_state[:10].Time.plot(kind= 'bar') #把次數最多的10個畫長方圖
```
