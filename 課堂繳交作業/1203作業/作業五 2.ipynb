```python
print("西元年份與天干地支對應")

year_list = list(range(2014, 2074))
sky_list = ['甲', '乙', '丙', '丁', '戊', '己', '庚', '辛', '壬', '癸']
land_list = ['子', '丑', '寅', '卯', '辰', '巳', '午', '未', '申', '酉', '戌', '亥']

for i in year_list:
    a = (i-1913) %10
    sky_str = sky_list[a]
    b = (i-1911) %12
    land_str = land_list[b]
    print('西元年', i, '天干地支為', sky_str + land_str)
```
