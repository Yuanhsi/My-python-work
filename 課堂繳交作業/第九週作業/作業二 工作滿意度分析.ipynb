# 勞工生活及就業狀況調查

## 資料來源：政府資料開放平台

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### 整體工作情形滿意度

```python
df = pd.read_csv('https://apiservice.mol.gov.tw/OdService/download/A17000000J-020078-LPO')
```

```python
df.head()
```

# 最滿意工作的前五名

```python
df_sort = df.sort_values('很滿意（%）', ascending = False)
```

```python
df_sort.head()
```

## 資料分群，看哪項因素對滿意度影響最大

```python
dfg= df.groupby(by = "統計項目別").sum()
```

```python
dfg_sort  = dfg.sort_values('很滿意（%）', ascending = False)
```

```python
dfg_sort
```

# 看一下收入對滿意度的影響

```python
dfm = df_sort[df_sort.統計項目別=='每月收入']
```

```python
dfm
```

```python
dfm_sort  = dfm.sort_values('很滿意（%）', ascending = False)
```

```python
dfm_sort.head()
```

# 發現收入未滿兩萬的人滿意度分居三四名，因此將很滿意與滿以加起來再看一次

```python
dfm["滿意與很滿意加總"]=dfm.iloc[:, 3:5].sum(axis=1)
dfm_sort  = dfm.sort_values('滿意與很滿意加總', ascending = False)
dfm_sort.head()
```
