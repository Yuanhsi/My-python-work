```python
import urllib.request as req
```

# 抓網頁原始碼

```python
url = "https://www.ptt.cc/bbs/movie/index.html"
```

```python
# 建立一個request物件，利用request 物件去打開網址

request = req.Request(url, headers={
    "User-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36"
}) 


with req.urlopen(request)as response:
    data=response.read().decode("utf-8") #取得網站原始碼（HTML,CSS,JS）
print(data)
```

# 解析資料，取得文章的標題

## 使用第三方套件 
## 在終端機中輸入 pip install beautifulsoup4

```python
import bs4 # 用beautifulsoup4去解析HTML
```

```python
root = bs4.BeautifulSoup(data, "html.parser")
```

```python
print(root.title.string) #抓取網頁標題
```

```python
titles = root.find("div", class_="title") # 用class去做篩選，find會找到其中一個符合條件的
```

```python
print(titles.a.string) # 只抓到其中一個標題
```

```python
titles = root.find_all("div", class_="title") #抓到所有的標題
```

```python
for title in titles:
    if title.a != None:
        print(title.a.string) # 如果標題有a，就印出來
```

# 連續抓取頁面的操作方式

## 1. 網站有cookie 的時候, 觀察cookie的作用

## 2. 想抓多頁的資料

```python
def getData(pageURL):

    request = req.Request(url, headers={
        "cookie":"over18=1",
        "User-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.108 Safari/537.36"
    }) 


    with req.urlopen(request)as response:
        data=response.read().decode("utf-8") #取得網站原始碼（HTML,CSS,JS）

    root = bs4.BeautifulSoup(data, "html.parser")
    titles = root.find_all("div", class_="title")

    for title in titles:
        if title.a != None:
            print(title.a.string)

    nextlink = root.find("a", string = "‹ 上頁") #找到‹ 上頁的a標籤
    return (nextlink["href"]) #要這個標籤的href屬性

    
```

```python
pageURL = "https://www.ptt.cc/bbs/Gossiping/index.html"
count = 0
while count<3:
    pageURL = "https://www.ptt.cc" + getData(pageURL)
    count += 1
print(pageURL)
```
